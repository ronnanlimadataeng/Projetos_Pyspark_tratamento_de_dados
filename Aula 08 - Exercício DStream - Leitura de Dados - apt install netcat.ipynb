2. Criar uma aplicação para ler os dados da porta 9999 e exibir no console

```python
#importando bibliotecas streaming e StreamingContext
from pyspark import SparkContext
from pyspark.streaming import StreamingContext
from time import sleep
```

```python
#conf para setar as configurações do spark
conf = SparkConf().setMaster("local[*]").setAppName("Dstream Python")
sc = SparkContext.getOrCreate(conf)
#criando sessão do spark(contexto do spark, de quantos em quantos segundos)
ssc = StreamingContext(sc,5)
```

```python
#ler estrutura(dsstream e excutar)
dstream = ssc.socketTextStream("localhost",9999)
```

```python
#printar estrutura
dstream.pprint()
```

```python
#startar aplicação - lembrar de utilizar o netcad para verificar as informações dentro do container
ssc.start() #parar executar novamente é necessario restart run no kernel
sleep(200) #teste para manter 20 segundos
```
